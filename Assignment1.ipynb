```python
import pandas as pd

df = pd.read_excel('Table1.ods')
df.head(n=6)

df.isna().sum(axis=0)

df['Course1'] = df['Course1'].fillna(df['Course1'].median()) 
df['Course3'] = df['Course3'].fillna(df['Course2'].median())

df.head()
```

```python
from sklearn import preprocessing

df2 = pd.read_excel('Table2.ods')
print(F"original:\n{df2}")

scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

result = scaler.fit_transform(df2[['Temperature', 'Precipitation', 'Humidity']])

print(F"result:\n{result}")

for i in range(3):
    df2.iloc[:,i + 1] = result[:, i]

df2.head(n=6)
```

```python
df3 = pd.read_excel("Table3.ods")
df3.head(n=6)

scaler3 = preprocessing.MinMaxScaler(feature_range=(0,1))

result = scaler3.fit_transform(df3[['Engine Size','Horsepower','Mileage','Price']])

print(F"result:\n{result}")

for i in range(4):
    df3.iloc[:, i + 3] = result[:, i]

df3['Age'] = 2023 - df3['Year']
df3.head()
```
